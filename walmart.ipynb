## Data Visualization 

```python
cnt_srs = walmart['Customer Segment'].value_counts()
plt.figure(figsize=(8,6))
sns.barplot(cnt_srs.index, cnt_srs.values, color='red',alpha=0.80)
plt.xlabel('Customer Segment', fontsize=12)
plt.ylabel('No.of Orders', fontsize=12)
plt.title('No.of Orders placed by various customer segment',fontsize=14)
plt.show()
```

```python
state = walmart['State'].value_counts()
state = state.head(10)
plt.figure(figsize=(10,8))
sns.barplot(state.index, state.values, color='Blue',alpha=0.70)
plt.xlabel('State', fontsize=12)
plt.ylabel('No.of Orders', fontsize=12)
plt.title('No.of Orders placed in a State',fontsize=14)
plt.show()
```

```python
category = walmart['Product Category'].value_counts()
plt.figure(figsize=(7,5))
sns.barplot(category.index, category.values, color='Yellow',alpha=0.70)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('No.of Orders', fontsize=12)
plt.title('No.of Orders that belongs to different Product Categories',fontsize=14)
plt.show()
```

```python
ship_mode = walmart['Ship Mode'].value_counts()
plt.figure(figsize=(7,5))
sns.barplot(ship_mode.index, ship_mode.values, color='magenta',alpha=0.70)
plt.xlabel('Ship Mode', fontsize=12)
plt.ylabel('No.of Orders', fontsize=12)
plt.title('No.of Orders shipped using different mode of transport',fontsize=14)
plt.show()
```

```python
containers = walmart['Product Container'].value_counts()
plt.figure(figsize=(10,8))
sns.barplot(containers.index, containers.values, color='cyan',alpha=0.70)
plt.xlabel('Product Container', fontsize=12)
plt.ylabel('No.of Orders', fontsize=12)
plt.title('Product Containers that are used while Orders',fontsize=14)
plt.show()
```

```python
priority = walmart['Order Priority'].value_counts()
plt.figure(figsize=(7,5))
sns.barplot(priority.index, priority.values, color='orange',alpha=0.90)
plt.xlabel('Order Priority', fontsize=12)
plt.ylabel('No.of Orders', fontsize=12)
plt.title('Priority wise Orders',fontsize=14)
plt.show()
```

```python
region = walmart['Region'].value_counts()
plt.figure(figsize=(7,5))
sns.barplot(region.index, region.values, color='teal',alpha=0.90)
plt.xlabel('Region', fontsize=12)
plt.ylabel('No.of Orders', fontsize=12)
plt.title('No.of Orders placed in a particular region',fontsize=14)
plt.show()
```

```python
date = (pd.DatetimeIndex(walmart['Order Date']).year).value_counts()
plt.figure(figsize=(7,5))
sns.barplot(date.index, date.values, color='green',alpha=0.90)
plt.xlabel('year', fontsize=12)
plt.ylabel('No.of Orders', fontsize=12)
plt.title('No.of Orders placed in an Year',fontsize=14)
plt.show()
```
